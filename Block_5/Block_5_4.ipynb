#### Область определения

```python
from sympy import Symbol, S
from sympy.calculus.util import continuous_domain

x = Symbol("x")
f = (x**3)/(x**2 - 1)
continuous_domain(f, x, S.Reals)
```

#### Область значений

```python
from sympy.calculus.util import function_range

x = Symbol("x")
f = (x**3)/(x**2 - 1)
function_range(f, x, S.Reals)
```

#### Точки пересечения с осями координат

```python
# Пересечение с осью Х
from sympy import solveset, Eq

f = (x**3)/(x**2 - 1)
solveset(Eq(f,0),x) # приравниваем функцию к нулю
```

```python
# Пересечение с осью Y
x = Symbol("x")
f = (x**3)/(x**2 - 1)
f.subs(x,0) # заменяем х на ноль
```

#### Исследование функции на четность и нечетность
#### Нахождение точек экстремума, областей возрастания и убывания

```python
from sympy import diff

x = Symbol("x")
expr = (x**3)/(x**2 - 1)
expr.diff(x)
```

```python
# Найдем точки в которых функция может менять знак
y = expr.diff(x)
solveset(Eq(y,0), x)
```
