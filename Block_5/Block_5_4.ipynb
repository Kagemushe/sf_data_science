#### <center>Операции над множествами в Python<center>

```python
A = {'ID453', 'ID312', 'ID41', 'ID3', 'ID500', 'ID920', 'ID36', 'ID27'}
B = {'ID41', 'ID36', 'ID27', 'ID124', 'ID7', 'ID501', 'ID91' }
```

##### Обьединение множеств

```python
union_AB = A.union(B)
print(union_AB)
```

```python
union_AB = A|B
print(union_AB)
```

##### Пересечение множеств

```python
inter_AB = A.intersection(B)
print(inter_AB)
```

```python
inter_AB = A&B
print(inter_AB)
```

##### Разность множеств

```python
diff_AB = A.difference(B)
print(diff_AB)
```

```python
diff_AB = A-B
print(diff_AB)
```

```python
diff_BA = B.difference(A)
print(diff_BA)
```

##### Симметрическая разность множеств

```python
symmAB = A.symmetric_difference(B)
print(symmAB)
```

```python
symmAB = A^B
print(symmAB)
```

#### <center>Библиотека Sympy, символьные вычисления и множества<center>

```python
from sympy import FiniteSet, Union
```

```python
l1 = [1, 3, 7, 10]
l2 = [9, 7, 1]
a = FiniteSet(*l1)
b = FiniteSet(*l2)
Union(a,b)
```

```python
from sympy import ConditionSet, Eq, Symbol, Interval

x = Symbol('x')
s = ConditionSet(x, Eq(x**2 - 5*x,0), Interval(2,9))
s
```
