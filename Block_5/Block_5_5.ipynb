### <center>1.Введение. Функции нескольких переменных<center>

```python
from sympy import diff, Eq, Symbol, S, solveset
```

```python
# Задание 1.3

x = Symbol("x")
expr = x**2 - 5*x + 4
y = expr.diff(x)
solveset(Eq(y,0),x)
```

```python
# Задание 1.4

x = Symbol("x")
expr = x**3 - 6*x**2 + 2
y = expr.diff(x)
solveset(Eq(y,0),x)
```

### <center>2.Частные производные<center>

```python
from sympy import symbols, cos, diff

a,b,c = symbols('a b c', real=True)
f = 5*a*b - a*cos(c) + a**8 + c**2*b
print(diff(f,a))
```

```python
# Задание 2.2

a,b,x = symbols('a b x', real=True)
f = 2*a*x + b**2
print(f.diff(a))
print(f.diff(a,2))
```

```python
# Задание 2.3

x,y = symbols('x y', real=True)
f = x**2 - y**2
print(f.diff(x))
print(f.diff(y))
```

```python
# Задание 2.4

a,b = symbols('a b', real=True)
f = (5-a-2*b)**2
print(f.diff(a))
print(f.diff(b))
```

```python
# Задание 2.5

x,y = symbols('x y', real=True)
f = x**2 + 3*x*y + 4*y**2
print(f.diff(x))
print(f.diff(y))
```

```python
# Задание 2.6

from sympy import sin

x,y = symbols('x y', real=True)
f = y*sin(x) + sin(y)
print(f.diff(x))
print(f.diff(y))
```

### <center>3.Безусловные экстремумы. Функции нескольких переменных<center>

```python
# Задание 3.2

from sympy import exp

x,y = symbols('x y', real=True)
f = exp(x) + 5*y**3*x

print(f.diff(x))
print(f.diff(y))
print('-'*30)
print(f.diff(x,2))
print(f.diff(x,y))
print(f.diff(y,2))
```
