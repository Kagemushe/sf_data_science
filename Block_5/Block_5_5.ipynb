```python
from sympy import diff, Eq, Symbol, S, solveset
```

```python
# Задание 1.3

x = Symbol("x")
expr = x**2 - 5*x + 4
y = expr.diff(x)
solveset(Eq(y,0),x)
```

```python
# Задание 1.4

x = Symbol("x")
expr = x**3 - 6*x**2 + 2
y = expr.diff(x)
solveset(Eq(y,0),x)
```
