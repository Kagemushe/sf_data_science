### <center>5.Практика: базовые действия над матрицами<center>

```python
Husband_Income = np.array([100,220,140])
Wife_Income = np.array([150,200,130])
Mother_In_Law_Income = np.array([90,80,100])

Husband_Consumption = np.array([50,50,60])
Wife_Consumption = np.array([100,80,140])
Mother_In_Law_Consumption = np.array([100,20,140])
```

```python
# Задание 5.1
inc_data = pd.DataFrame([Husband_Income, Wife_Income, Mother_In_Law_Income])
inc_matrix = inc_data.values
inc_matrix = inc_matrix.T
inc_matrix
```

```python
# Задание 5.2
cons_data = pd.DataFrame([Husband_Consumption, Wife_Consumption, Mother_In_Law_Consumption])
cons_matrix = cons_data.values
cons_matrix = cons_matrix.T
cons_matrix
```

```python
# Задание 5.3
bill = 0.13*inc_matrix
after_bill = inc_matrix - bill
after_bill[0]
```

```python
# Задание 5.4
P = after_bill - cons_matrix
P[2]
```
