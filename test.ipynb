```python
def experience_work(x):
    """Функция принимает признак с описание места и опыта работы,
    и возвращает числовой признак опыта работы в месяцах.

    Args:
        x (Признак в Data Frame): Признак с полным описние места работы и опыта работы в 
        годах и/или месяцах.

    Returns:
        int: Числовой признак опыта работы в месяцах.
    """
    if x is np.nan or x == 'Не указано': # Находим пустые и скрытые пустые значения и заменяем их None
        return None
    
    years = ['год', 'года', 'лет'] # Вспомогательный список слов опыта в годах
    mounths = ['месяц', 'месяцев', 'месяца'] # Вспомогательный список слов опыта в месяцах
    years_exp = 0
    mounths_exp = 0
    x_splited = x.split(' ')[:7] # Разделяем строку и оставляем первые 6 элементов с интересующей нас информацией
    
    for index, elem in enumerate(x_splited):
        if elem in years: # Если элемент есть в списке years, преобразуем предыдущий элемент опыта в годах в число
            years_exp = int(x_splited[index-1])
        if elem in mounths: # Если элемент есть в списке mounths, преобразуем предыдущий элемент опыта в месяцах в число
            mounths_exp = int(x_splited[index-1]) 
                 
    return (years_exp*12 + mounths_exp) # Возвращаем суммарный опыт работы
```

```python
# Создаем новый признак, применяя нашу функцию к имеющимуся признаку "Опыт работы"
hh_data['Опыт работы(месяц)'] = hh_data['Опыт работы'].apply(experience_work)
```

```python
def cities(y):
    """Функция принимает на вход признак Data Frame и возвращает информацию только по городам
    проживания соискателей.

    Args:
        y (Признак в Data Frame): Признак включающий в себя информацию о городе проживания, метро,
        готовности к переезду и готовности к командировкам.

    Returns:
        Строковое название города: Возвращает 4 категории: Москва, Санкт-Петербург, город-миллионник, другие.
    """
    y_splited = y.split(' , ') # Разделяем строку по запятой с пробелами
    grand_cities = ['Москва', 'Санкт-Петербург'] # Вспомогательний список крупнейших городов-категорий
    # Вспомогательний список городов-миллионников
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    if y_splited[0] in grand_cities:
        return y_splited[0]
    elif y_splited[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

# Создаем новый признак "Город", применяя функцию к исходному признаку   
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(cities)    
```

```python
def get_moving(y):
    """Функция принимает на вход признак Data Frame и возвращает булевые значения готовности к перезду.

    Args:
        y (Признак в Data Frame): Признак включающий в себя информацию о городе проживания, метро,
        готовности к переезду и готовности к командировкам.

    Returns:
        bool: Булевое значение только по готовности к переезду.
    """
    y_splited = y.replace(' ', '').split(',') # Убираем пропуски и делим строку по ","
    if len(y_splited) > 3: # Если длина списка больше 3, тоесть присуствует указание о станции метро
        if y_splited[2][0:2] == 'не': # сравниваем 2 первых знака третьего элемента в списке
            return False              # готовность к переезду начинается с 'не'
        else:
            return True
    else:
        if y_splited[1][0:2] == 'не':
            return False
        else:
            return True    

# Создаем новый признак "Готовность к переезду", применяя функцию к исходному признаку        
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_moving)      
```

```python
def get_buissnes_trip(y):
    """Функция принимает на вход признак Data Frame и возвращает булевые значения готовности к командировкам.

    Args:
        y (Признак в Data Frame): Признак включающий в себя информацию о городе проживания, метро,
        готовности к переезду и готовности к командировкам.

    Returns:
        bool: Булевое значение только по готовности к командировкам.
    """
    y_splited = y.replace(' ', '').split(',') # Убираем пропуски и делим строку по ","
    if len(y_splited) > 3: # Если длина списка больше 3, тоесть присуствует указание о станции метро
        if y_splited[3][0:2] == 'не': # сравниваем 2 первых знака четвертого элемента в списке
            return False              # готовность к командировкам начинается с 'не'
        else:
            return True
    else:
        if y_splited[-1][0:2] == 'не':
            return False
        else:
            return True   
        
# Создаем новый признак "Готовность к командировкам", применяя функцию к исходному признаку
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_buissnes_trip) 
```

```python
#Создаем функцию, разделяющую строку по запятой и пробелу
func_splited = lambda x: x.split(', ')
# Применяем ее к признакам "Занятость" и "График"
hh_data['Занятость'] = hh_data['Занятость'].apply(func_splited)
hh_data['График'] = hh_data['График'].apply(func_splited)

# определяем уникальные категории для интересующих нас признаков, и создаем из них список
metrics_employment = hh_data['Занятость'].value_counts().sort_index(key=lambda x: x.str.len(), ascending=False).index[0]
metrics_schedule = hh_data['График'].value_counts().sort_index(key=lambda x: x.str.len(), ascending=False).index[0]

# Создаем новые признаки из исходного признака "Занятость"
# Новые признаки будут содержать в себе значения True и False
for employment in metrics_employment:
    hh_data[employment] = hh_data['Занятость'].apply(lambda x: employment in x if x else False)

# Создаем новые признаки из исходного признака "График"
# Новые признаки будут содержать в себе значения True и False
for schedule in metrics_schedule:
    hh_data[schedule] = hh_data['График'].apply(lambda x: schedule in x if x else False)
```
