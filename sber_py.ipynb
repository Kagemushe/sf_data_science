```python
# Задание 2.4
fig = plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(
    data=sber_data,
    x='ecology',
    y='price_doc',
    orient='v',
    width=0.9
)
boxplot.set_title('Зависимость цены от уровня экологии района')
boxplot.set_xlabel('Экологический уровень')
boxplot.set_ylabel('Показатель цены')
```

```python
# Задание 2.5
fig = plt.figure(figsize=(10, 7))
relplot = sns.relplot(
    data=sber_data,
    x='kremlin_km',
    y='price_doc',
    size='price_doc'
)
relplot.set(xscale='log', yscale='log')
relplot.set_titles('Зависимость цены от расстояния до центра Москвы')
relplot.set_xlabels('Расстояние до центра')
relplot.set_ylabels('Цена')
```

```python
display(sber_data.isnull().tail())
```

```python
cols_null_percent = sber_data.isnull().mean() * 100
cols_with_null = cols_null_percent[cols_null_percent>0].sort_values(ascending=False)
display(cols_with_null)
```

```python
cols_with_null.plot(
    kind='bar',
    figsize=(10, 4),
    title='Распределение числа пропусков'
);
```

```python
colors = ['blue', 'yellow']
fig = plt.figure(figsize=(10, 4))
cols = cols_with_null.index
ax = sns.heatmap(
    sber_data[cols].isnull(),
    cmap=sns.color_palette(colors),
)
```

```python
# создаем копию исходной таблицы
drop_data = sber_data.copy()
# задаем минимальный порог: вычисляем 70% от числа строк
thresh = drop_data.shape[0]*0.7
# удаляем столбцы в которых более 30% пропусков
drop_data = drop_data.dropna(thresh=thresh, axis=1)
# удаляем строки, в которых усть хотя бы один пропуск
drop_data = drop_data.dropna(how='any', axis=0)
# отображаем результирующую долю процентов
drop_data.isnull().mean()
```
